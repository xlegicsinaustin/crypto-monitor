```python
import pandas as pd 
import sqlalchemy as sql
import unicorn_binance_websocket_api as unicorn
```

```python
#defining engine
engine = sql.create_engine('sqlite:///Liveprice.db')
```

```python
df = pd.read_sql('MATICUSDT',engine)
df
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
```

```python
def animate(i):
    data = pd.read_sql('MATICUSDT',engine)
    plt.cla() #clears the axis on refresh
    plt.plot(data.timestamp,data.price, color='g') #price
    plt.axhline(y=data.price.mean(),color='b') #rolling mean
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.title('MATICUSDT')
    plt.gcf().autofmt_xdate()
    plt.tight_layout()
```

```python
sns.set_style('darkgrid')
```

```python
ani = animation.FuncAnimation(plt.gcf(),animate,100)
plt.tight_layout()
plt.show()
```
